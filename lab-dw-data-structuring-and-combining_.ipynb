df2_clean:

```python
import cleanfunctions

# correct 'state' AZ to Arizona
df2_clean['state'] = df2_clean['state'].apply(lambda x: 'Arizona' if str(x) == 'AZ' else str(x))
df2_clean['state'].value_counts()

# clean gender column
df2_clean = cleanfunctions.clean_gender_column(df2_clean)
df2_clean['gender'].value_counts()


# correct 'education' Bachelors --> Bachelor
df2_clean['education'] = df2_clean['education'].apply(lambda x: 'Bachelor' if str(x) == 'Bachelors' else str(x))
df2_clean['education'].value_counts()

# clean 'customer_lifetime_value'
df2_clean['customer_lifetime_value'] = df2_clean['customer_lifetime_value'].apply(lambda x: str(x).replace('%','') if ('%' in str(x)) else str(x))
df2_clean['customer_lifetime_value'] = df2_clean['customer_lifetime_value'].astype('float64')

# clean 'number_of_open_complaints'
df2_clean['number_of_open_complaints'] = df2_clean['number_of_open_complaints'].apply(lambda x: int(str(x).split('/')[1] if pd.notnull(x) else x))
display(df2_clean['number_of_open_complaints'].value_counts()) # check counts and unique values after correcting

# clean 'vehicle_class': to luxury
df2_clean['vehicle_class'] = df2_clean['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Sports Car', 'Luxury SUV', 'Luxury Car'] else x)
df2_clean['vehicle_class'].value_counts()

# based on df2_clean.info, there are 5 records with null value for
# column 'gender' which was converted to value 'U' as unknown
# and another 4 records with null value in column 'customer_lifetime_value'
# which will be dropped (as it is 4 records out of total 996 records)

# drop na
df2_clean = df2_clean.dropna()
df2_clean = df2_clean.reset_index()

# dop column 'index'
df2_clean = df2_clean.drop(columns='index')
df2_clean

# convert all numeric columns to integer
df2_clean = df2_clean.applymap(lambda x: int(x) if isinstance(x, (float,int)) else x)
display(df2_clean.dtypes)

                                    
```

```python
%%writefile -a '/Users/minhnguyen/IronHack2023-2024/Bootcamp/Week_1/Day_3/Afternoon/lab-dw-data-cleaning-and-formatting/cleanfunctions.py'

def check_value_count_each_column (df:pd.DataFrame):
    '''
    This function takes the input pandas dataframe, 
    then print value_counts for each column in the dataframe
    Input:
    df: Pandas dataframe
    Output:
    none
    '''
    for cols in list(df.columns):
        print('\nValue counts of',cols,'\n',df[cols].value_counts())

```

```python
# 2. need to delete unnamed:_0 column

marketing_customer_df_clean = marketing_customer_df.drop(columns='unnamed:_0')
marketing_customer_df_clean
```

```python
# 3. column 'effective_to_date' should be in datetime format, yyyy-mm-dd

marketing_customer_df_clean['effective_to_date'] = pd.to_datetime(marketing_customer_df['effective_to_date'], errors='coerce')
display(marketing_customer_df_clean)       

#print(marketing_customer_df.dtypes)
print(marketing_customer_df_clean.dtypes)
```

```python
# 4. column 'months_since_last_claim' (col[14]) --> dtype change to int64
#marketing_customer_df['months_since_last_claim']= marketing_customer_df['months_since_last_claim'].applymap(lambda x:int(x) if isinstance (x, (float,int)) else x) 
marketing_customer_df_clean['months_since_last_claim']= marketing_customer_df['months_since_last_claim'].round().astype('int64') 

display(marketing_customer_df_clean['months_since_last_claim'].value_counts())
#display(type(marketing_customer_df['months_since_last_claim']))

#print(marketing_customer_df.dtypes)
print(marketing_customer_df_clean.dtypes)

display(marketing_customer_df_clean.head())
```

```python
#5. column 'number_of_open_complaints' (col[16]) 
#    * the current dtype is float64. This should be int64, not a float64 number as you can not have a half of a complaint.
#    * considering the nature of the information of this column, the amount (633 records, is about 5.8% of total records),the approach I will take is to convert this colnumn to int64, and round the float value to the nearest int value.
marketing_customer_df_clean['number_of_open_complaints']= marketing_customer_df['number_of_open_complaints'].round().astype('int64') 

display(marketing_customer_df_clean['number_of_open_complaints'].value_counts())

#print(marketing_customer_df.dtypes)
print(marketing_customer_df_clean.dtypes)
# Caution: after this data type convertion, the number of O complaints
# goes from 8160 to 8793 records. This might influence the interpretation of data.

display(marketing_customer_df_clean.head())
```

```python
#6. change column name 'employmentstatus' to 'employment_status' for the consistency of column name
marketing_customer_df_clean=marketing_customer_df_clean.rename(columns = {'employmentstatus': 'employment_status'})
marketing_customer_df_clean
```

```python
#7. values in 'vehicle_class': 'Sport cars', 'luxury SUV', and 'luxury' could be replaced by 'luxury'
marketing_customer_df_clean['vehicle_class']=marketing_customer_df_clean['vehicle_class'].apply(lambda x: 'Luxury' if x in ['Sports Car', 'Luxury SUV', 'Luxury Car'] else x) 
display(marketing_customer_df_clean['vehicle_class'].value_counts())
```

```python
#8. 'vehicle_size': replace 'medsize' with 'Medium'
marketing_customer_df_clean['vehicle_size'] = marketing_customer_df_clean['vehicle_size'].apply(lambda x: 'Medium' if x == 'Medsize' else x )
display(marketing_customer_df_clean['vehicle_size'].value_counts())
```
